## Graph Visualization

There is some opportunity to use fmdtools.graph to visualize prognostics models. Here we show what this could look like with the Powertrain composite model.

```python
from prog_models.models import Powertrain, ESC, DCMotor
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
```

```python
esc = ESC()
motor = DCMotor()
powertrain = Powertrain(esc, motor)
```

The goal of this experiment will be to show the input/output structure of ProgPy models. Unfortunately, it's not clear that there is a property containing the name of an object, so we'll have to pass the name instead:

```python
def make_graph(structure, structurename):
    """
    Makes a networkx digraph of a ProgPy model with the given structure name.
    
    Parameters
    ----------
    structure: ProgPy model
    structurename: str (name of the model)
    Returns
    -------
    networkx digraph with structurename as the central node and inputs/outputs 
    as nodes with edges corresponding to the input/output property
    """
    g = nx.DiGraph()
    g.add_node(structurename, bipartite=0) #note the bipartite parameter: used to distinguish component (0) vs values (1)
    for val in structure.inputs:
        g.add_node(val, bipartite=1)
        g.add_edge(val,structurename)
    for val in structure.outputs:
        g.add_node(val, bipartite=1)
        g.add_edge(structurename,val)
    return g
```

```python
g_esc = make_graph(esc, "esc")
g_esc
```

We can visualize this graph with nx.draw (though there are certainly other tools that can visualize networkx graphs). Below we show this in a basic function. Note the layout parameter.

```python
def draw_graph(g, pos ={}, layout = "kamada_kawai_layout"):
    """
    Draws a graph g with labels in a given layout (unless a dictionary of positions pos is provided)
    Parameters
    ----------
    g : graph
    pos : dict of node positions
    layout : layout to use if no pos (see: https://networkx.org/documentation/stable/reference/drawing.html)
    Returns
    ----------
    fig : figure 
    """
    fig = plt.figure()
    if not pos: pos = getattr(nx.layout, layout)(g)
    nx.draw(g, pos=pos)
    nx.draw_networkx_labels(g, pos)
    return fig
```

```python
fig = draw_graph(g_esc)
```

```python
import fmdtools.resultdisp as rd
```

We can similarly plot this in fmdtools. Note that the default renderer is the same was we were using above, just with some more stylistic details (and parameters we can change).

```python
rd.graph.show(g_esc)
```

```python
rd.graph.show(g_esc, highlight=[["esc"],[],[]])
```

We can also use some other renderers (provided they are installed), such as graphviz, using the renderer parameter.

These graphs can further be useful for composite models. Unfortunately, there isn't much provided in the class structure to show how this model is composed of components. The following function thus creates a networkx digraph structure given a list of component names and just assumes they are connected via similar-named values:

```python
def make_compositegraph(compmodel, sub_names):
    """
    Makes a networkx digraph of a ProgPy composite model with the given sub-models sub_names
    
    Parameters
    ----------
    compmodel: ProgPy composite model
    sub_names: list of str (name of the model)
    Returns
    -------
    networkx digraph with components and shared values as nodes
    """
    g=nx.DiGraph()
    for name in sub_names:
        sub_structure = getattr(compmodel, name)
        sub_g = make_graph(sub_structure, name)
        g = nx.compose(g, sub_g)
    return g
```

```python
g_powertrain = make_compositegraph(powertrain, ["esc", "motor"])
```

```python
fig=draw_graph(g_powertrain)
```

```python
%matplotlib qt
pos = rd.graph.set_pos(g_powertrain)
```

### Discussion

There are a few things we can surmise from these experiments if we want to have the ability to use fmdtools to visualize ProgPy models.
- It would be helpful if ProgPy models had more attributes which indicated their structure. Specifically:
    - Models should have a "name" attribute to distinguish different components/models
    - Composite Models should have a list of components 
    - Models should have some structured way of mapping the inputs of one 
    - With the following in place, we can make a graph structure. This method could be a part of the model structure (or otherwise some glue code), along with a show() method
- More capabilities should be developed in fmdtools to generalize to more classes of models. Specifically:
    - It should be easy to change/modify all graph attributes (e.g., arrows)
    - We may want to add more types of visualizations (e.g. listing values on the plot, etc)
    - It may be helpful to move some specialized fmdtools boilerplate model code out of graph and into somewhere else (perhaps the model structure itself)
